```python
import numpy as np
```

```python
# 1D array
a = np.random.randint(2, 10, 5)
b =  np.random.randint(10, 24, 5)
c =  np.random.randint(30, 40, 5)

# 2D array

m1 = np.random.randint(2, 40, size = (5, 6))
m2 = np.random.randint(10, 50, size = (5, 6))
m3 = np.random.randint(30, 120, size = (5, 6))

print(m1)
print(m2)
print(m3)
```

```python
# Addition
add_arr = a+b
print(add_arr)

add_arr1 = np.add(a, b)
print(add_arr1)

# for 2D array
print(np.add(m1, m2))
```

```python
# for three or more array
add_arr2 = np.add( a, b, c)
print(add_arr2)

# 2D
print(np.add(m1, m2, m3))
```

```python
# Subtraction
sub1 = b-a
print(sub1)

sub2 = np.subtract(b, a) # b-a
print(sub2)

# for 2D
print(np.subtract(m1, m2))

```

```python
# for more than two array 
# let's calculate for (c-b-a)

sub3 = np.subtract(c, b, a)
print(sub3)

# for 2D
print(np.subtract(m3, m2, m1)) # m3 - m2- m1

# Also we can subtract a number from every element 
sub4 = np.subtract(a, 1)
print(sub4)
```

```python
# multiplication ( element wise)
# eg: a =[a1, a2, a3]  and b = [b1, b2, b3]
# element wise multiplication is  [a1*b1, a2*b2, a3*b3]

mult1 = a*b
print(mult1)

mult2 = np.multiply(a, b)
print(mult2)

# for 2D
print(np.multiply(m1, m2))
```

```python
# Dot multiplication
# a = [a1, a2, a3]  --> shape(a) = (1,3)
# b = [b1,
#      b2, 
#      b3]           --> shape(b) = (3, 1)

# So, resulting matrix is of shape (1, 1)
a.shape = (1, 5)
 
b.shape = (1, 5)

print(b.T.shape)


dot_m1 = np.dot(a, b.T)
print(dot_m1)

# Another code for dot 
print(a@b.T)

# for 2D
print(np.dot(m1, m2.T))
```

```python
import csv 
    
# a rows as dictionary objects 
mydict =[{'branch': 'COE', 'cgpa': '9.0', 'name': 'Nikhil', 'year': '2'}, 
         {'branch': 'COE', 'cgpa': '9.1', 'name': 'Sanchit', 'year': '2'}, 
         {'branch': 'IT', 'cgpa': '9.3', 'name': 'Aditya', 'year': '2'}, 
         {'branch': 'SE', 'cgpa': '9.5', 'name': 'Sagar', 'year': '1'}, 
         {'branch': 'MCE', 'cgpa': '7.8', 'name': 'Prateek', 'year': '3'}, 
         {'branch': 'EP', 'cgpa': '9.1', 'name': 'Sahil', 'year': '2'}] 
    
fields = ['name', 'branch', 'year', 'cgpa'] 
    
filename = "university_records.csv"
     
with open(filename, 'w') as csvfile: 
  
    writer = csv.DictWriter(csvfile, fieldnames = fields) 
        
    writer.writeheader() 
        
    writer.writerows(mydict) 
```

```python
with open('university_records.csv', 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  for i in csv_reader:
    print(i)
```
