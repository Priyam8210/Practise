```python
# concept of grid

x_pt = np.array([2, 3, 5, 7, 9, 12, 8, 17, 15])
y_pt = np.array([6, 16, 2, 24, 29, 44, 5, 33, 44])

plt.plot(x_pt, y_pt, c = 'green', linewidth = '5')

x_pt = np.array([2, 3, 5, 4, 7, 9, 12])
y_pt = np.array([6, 16, 2, 19, 24, 29, 44])

plt.plot(x_pt, y_pt, c = 'orange', linewidth = '2') 
plt.xlabel("number of mens")                                   # label for x- axis
plt.ylabel("working hours")                                   # label for y- axis
plt.title("working comparision")                              # Title for graph

plt.grid()       # <-----

plt.show()
```

```python
# vertical plotting

x_pt = np.array(["a", "b", "c", "d", "e", "f"])
y_pt = np.array([24, 12, 17, 19, 34, 26])

plt.bar(x_pt, y_pt, color = "green", width = .2)
plt.show()
```

```python
# horizontal plotting

x_pt = np.array(["a", "b", "c", "d", "e", "f"])
y_pt = np.array([24, 12, 17, 19, 34, 26])

plt.barh(x_pt, y_pt)
plt.show()
```

```python
x_pt = np.array(["a", "b", "c", "d", "e", "f"])
y_pt = np.array([24, 12, 17, 19, 34, 26])

plt.barh(x_pt, y_pt, color = "orange", height = .3)
plt.show()
```

```python
x_pt = np.array(["a", "b", "c", "d", "e", "f"])
y_pt = np.array([24, 12, 17, 19, 34, 26])

plt.pie(y_pt, labels = x_pt)
plt.show()
```

```python
x_pt = np.array(["a", "b", "c", "d", "e", "f"])
y_pt = np.array([24, 12, 17, 19, 34, 26])
my_exp = [.1, 0, .5, 0, .5, 0]

plt.pie(y_pt, labels = x_pt, explode = my_exp,  shadow = True)
plt.legend(title = "Mans label")
plt.show()
```
